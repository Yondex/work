```python
#--Компьютерная фирма: Вывести список всех продуктов и производителя с указанием типа продукта 
#(pc, printer, laptop). Вывести: model, maker, type
request = """
select pr.model, maker, pr.type from product pr join laptop lp on pr.model=lp.model 
union
(select pr.model, maker, pr.type from product pr join pc on pr.model=pc.model )
union
(select pr.model, maker, pr.type from product pr join printer on pr.model=printer.model )

"""
pd.read_sql_query(request, conn)
```

```python
#--task14 (lesson3)
#--Компьютерная фирма: При выводе всех значений из таблицы printer дополнительно вывести для тех, 
#у кого цена вышей средней PC - "1", у остальных - "0"
request = """
SELECT *,  
    CASE WHEN price > (select avg(price) from Printer) THEN '1'
         ELSE '0'
    END
from Printer
"""
pd.read_sql_query(request, conn)
```

```python
#--task15 (lesson3)
#--Корабли: Вывести список кораблей, у которых class отсутствует (IS NULL)
# В Таблице ships все корабли с классами. В таблице outcomes есть корабли которых нет в таблице ships
request = """
SELECT ship FROM outcomes
except 
SELECT name FROM ships

"""
pd.read_sql_query(request, conn)
```

```python
#--task16 (lesson3)
#--Корабли: Укажите сражения, которые произошли в годы, не совпадающие ни с одним из годов спуска кораблей на воду.
request = """

SELECT name FROM battles
where cast(extract(year from date) AS int) not in (select launched from ships)

"""
pd.read_sql_query(request, conn)
```

```python
#--task17 (lesson3)
#--Корабли: Найдите сражения, в которых участвовали корабли класса Kongo из таблицы Ships.
request = """

SELECT battles.name FROM outcomes join ships on outcomes.ship=ships.name
join battles on outcomes.battle = battles.name
where class='Kongo'
"""
pd.read_sql_query(request, conn)
```

```python
#--task1  (lesson4)
#-- Компьютерная фирма: Сделать view (название all_products_flag_300) для всех товаров (pc, printer, laptop)
#с флагом, если стоимость больше > 300. Во view три колонки: model, price, flag


request = """
create view all_products_flag_300 as
(
 select model, price, 'PC' from pc where price>300
 union all
 select model, price, 'Printer' from Printer where price>300
 union all
 select model, price, 'Laptop' from Laptop where price>300

)
"""
pd.read_sql_query(request, conn)
```

```python
#--task3  (lesson4)
#-- Компьютерная фирма: Вывести все принтеры производителя = 'A' 
#со стоимостью выше средней по принтерам производителя = 'D' и 'C'. 
#Вывести model

request = """
select product.model from printer join product on printer.model = product.model
where product.maker='A' and printer.price >(select avg(price) from printer join product on printer.model = product.model
where product.maker='D' ) and printer.price > (select avg(price) from printer join product on printer.model = product.model
where product.maker='C' )

"""
pd.read_sql_query(request, conn)
```
