```python
vacs_id = []
vacancy_ids = []
for i in tqdm(range(vacancies.get('pages'))):
    url = f'https://api.hh.ru/vacancies?page={i}&per_page={num_per_page}&area={moscow}'
    res = requests.get(url)
    vacancies = res.json()  
    vacancy_ids = [el.get('id') for el in vacancies.get('items')]
    vacs_id.extend(vacancy_ids)
```
