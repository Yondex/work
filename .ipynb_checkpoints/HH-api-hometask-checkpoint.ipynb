```python
# Вытащите все описания этих вакансий
vacancies = res.json()
num_pages = vacancies.get('pages')

obch = []
for par in range(1,num_pages+1):
    url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}'
    res = requests.get(url)
    vacancies = res.json()
    snippet_r = [el.get('snippet') for el in vacancies.get('items')] 
    for elem in snippet_r:
        df=elem.get('responsibility')
        sds=df
        obch.append(sds)
    page=par
obch
```

```python
vacs = vacancies['items']
df = pd.DataFrame(vacs)
df['skills']=''
for el in vacancies.get('items'):
    id_vacs=el.get('id')
    url = f'https://api.hh.ru/vacancies/39932897?host=hh.ru'
    res = requests.get(url)
    skills = res.json()
    s= skills.get('key_skills')
    if s==None:
        s=0
    df['skills']=s
```
