```python
import requests
import json
from tqdm.auto import tqdm
from collections import defaultdict, Counter
import pickle 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime, date, time
import itertools
```

```python
sber = '3529'
page = 1
num_per_page = 100
moscow = 113
id_vacs = 1
url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}'

pd.set_option('display.max_columns', 30) 
```

```python
res = requests.get(url)
```

```python
vacancies = res.json()
num_pages = vacancies.get('pages')
```

```python
# Вытащите все описания этих вакансий
vacs_id = []
vacancy_ids = []
#for i in tqdm(range(vacancies.get('pages'))):
try:
    for i in tqdm(range(1, 100)):
        url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={i}&per_page={num_per_page}&area={moscow}'
        res = requests.get(url)
        vacancies = res.json()  
        vacancy_ids = [el.get('id') for el in vacancies.get('items')]
        vacs_id.extend(vacancy_ids)
        k = i
        
except:
    del vacs_id[:]
    for j in tqdm(range(0, k)):
        url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={j}&per_page={num_per_page}&area={moscow}'
        res = requests.get(url)
        vacancies = res.json()
        vacancy_ids = [el.get('id') for el in vacancies.get('items')]
        vacs_id.extend(vacancy_ids)
```

```python
vacan = defaultdict(list)
for i in tqdm(vacs_id):
    url = f'https://api.hh.ru/vacancies/{i}'
    res = requests.get(url)
    query = res.json()
    vacan['id'].append(query.get('id'))
    vacan['published'].append(query.get('published_at'))
    vacan['name'].append(query.get('name'))
    vacan['address'].append(query.get('address'))
    vacan['description'].append(query.get('description'))
    vacan['key_skills'].append(query.get('key_skills'))
  
    
df = pd.DataFrame(vacan)
df
    
```

```python
sk = vacan.get('key_skills')
vacan['key_skills'] = [',' .join([y.get('name') for y in x]) for x in sk]
```

```python
df = pd.DataFrame(vacan)
df.head(50)
```

```python
# Переведите даты публикаций в datetime
df["published"] = pd.to_datetime(df["published"],errors="coerce").dt.strftime("%d-%m-%Y")
```

```python
# Постройте график опубликованных вакансий по датам
df.groupby('published')['id'].count().plot(kind='barh');
```

```python
# Переведите даты в дни недели, и определите день недели, в который больше всего публикуют вакансий
df['dayofweek']= pd.to_datetime(df['published']).dt.dayofweek
```

```python
df.groupby('dayofweek')['id'].count().plot(); # В срезу публикуют больше вакансии чем в остальные дни недели(0 - monday, 1 - tuesday ..)
```

```python
#Найдите те вакансии с использованием python, которые вам интересны (искал по ключевому слову 'data')
# Определите по полю skills какие навыки больше всего востребованы для этих вакансий, и
# Постройте график наиболее востребованных вакансий
name2 ='data'
n2 = df.name.str.lower().str.contains(name2)
l = list(filter(None, df[n2].key_skills.tolist()))
l = ','.join(l).split(',')
count_skills = Counter(l)
j = count_skills.most_common(10)
d = pd.DataFrame(j)
d.rename(columns={0: 'skills', 1: 'count'}, inplace=True)
d.plot(x="skills", y="count", kind="bar", rot=100, fontsize=20);
```
