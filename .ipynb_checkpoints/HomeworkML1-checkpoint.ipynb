```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn import tree
import graphviz
```

```python
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV',axis=1), df['MEDV'], test_size=0.25, random_state=54)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.fit
```

```python
pred=reg.predict(X_test) 
pred
```

```python
y_train.describe()
```

```python
print(mean_squared_error(y_test, pred, squared=True))
```

```python
plt.figure(figsize=(20,9))
plt.plot(y_test.values, "blue",linewidth=5, label='Реальные значения', alpha=0.5)
plt.plot(pred, "red",linewidth=5, label='Предсказание')
plt.legend(loc='best')
plt.title("График пример")
plt.grid(True)
plt.show();
```

```python
clf = tree.DecisionTreeRegressor()
```

```python
clf = clf.fit(X_train, y_train) 
```

```python
clf.predict(X_test[-1:].values)
```

```python
red_clf = clf.predict(X_test)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render('smpl') 
```

```python
#строит дерево
dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=df.drop('MEDV',axis=1).columns,  
                      class_names=df['MEDV'],  
                      filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```

```python
clf.feature_importances_
```

```python
frame = { 'FeatureImportance': clf.feature_importances_, 'Feature': X_train.columns }
result = pd.DataFrame(frame)
result.sort_values(by='FeatureImportance', ascending=False).head(10)
```
