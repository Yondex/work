```python
import csv
import sys
from datetime import datetime, timedelta
csv.field_size_limit(sys.maxsize)
```

```python
vacs = list(csv.DictReader(open('Vacancy.csv')))
```

```python
# Сколько вакансий, которые вам нравятся?
count = 0
for i in range(len(vacs)):
    dic = vacs[i]
    vactitle=dic.get('vactitle').lower() # В переменную vactitle получаем значение поля 'vactitle' в нижнем регистре
    k = vactitle.rfind('дизайнер')  # rfind возвращает последний индекс если найдена строка
    if k>=0:                         # Если значения к больше или равно 0 тогда подстрока найдена
        count+=1
print(count)
```

```python
# Сколько вакансий для аналатика данных?
count = 0
for i in range(len(vacs)):
    dic = vacs[i]
    vactitle=dic.get('vactitle').lower() # В переменную vactitle получаем значение поля 'vactitle' в нижнем регистре
    k = vactitle.rfind('аналитик')  # rfind возвращает последний индекс если найдена строка
    if k>=0:                         # Если значения к больше или равно 0 тогда подстрока найдена
        count+=1
print(count)
```

```python
# Сколько вакансий для аналитика данных с использованием Python?
count = 0
for i in range(len(vacs)):
    dic = vacs[i]
    vactitle=dic.get('vactitle').lower() # В переменную vactitle получаем значение поля 'vactitle' в нижнем регистре
    vacdescription =dic.get('vacdescription').lower()  #В переменную vacdescription получаем значение поля 'vacdescription' в нижнем регистре
    vt = vactitle.rfind('аналитик')  # rfind возвращает последний индекс если найдена строка
    vd = vacdescription.rfind('python')
    if (vt>=0 and vd>=0):            # Если значения vt и vd больше или равно 0 тогда подстроки найдены
        count+=1
print(count)
```

```python
# Сколько вакансий с позициями на которых вы работаете?
count = 0
for i in range(len(vacs)):
    dic = vacs[i]
    vactitle=dic.get('vactitle').lower()  # В переменную vactitle получаем значение поля 'vactitle' в нижнем регистре
    k = vactitle.rfind('инженер')  # rfind возвращает последний индекс если найдена строка
    if k>=0:                       # Если значения к больше или равно 0 тогда подстрока найдена
        count+=1
print(count)
```

```python
# Насколько свежие эти вакансии?
# Дата создания всех записей 14.03, Дата изменений всех записей 15.03 - из-за этого получается 2 ответа либо 0 либо 4002
now = datetime.now()
count = 0
for i in range(len(vacs)):
    dic = vacs[i]
    vactitle=dic.get('vactitle')
    vacstatus=dic.get('created_at')
    date_time_str = vacstatus
    date_time_obj = datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S.%f')
    td = timedelta(30) # Дельта даты выбрал 30 дней.
    n = datetime.now()  # Текущая дата
    q = n-td            # q= текущую дату отнимаем 30 дней и сравниваем с датой созданием вакансии
    if q>date_time_obj:
        count+=1
print(count)
```
